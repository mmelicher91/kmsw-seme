
1. **Polynom ((x+1)^2) / x**






```python
x = sympy.Symbol("x")
sympy.integrate((2*x*(x+1)-(x+1)**2)/(x**2),x)
```

```python
def f(x): 
    return ((2*x*(x+1)-(x+1)**2))/(x**2)
   
a = 1
b = 4
dx = 0.1
```

Integrace pomocí lichoběžníků

```python
integral = 0
x = a
i = 0
while x < b:
    integral += dx * (f(x) + f(x+dx))/2
    x += dx
print(integral)
```

Integrace ve formě Newton-Cotes vzorec

```python
n = int((b-a)//dx)+1
integral = f(a) + f(b)
for i in range(1, n):
    integral += 2*f(a+i*dx)
integral *= dx/2
print(integral)
```

Integrace pomocí knihovny numpy

```python
x = np.arange(a, b+dx, dx)
y = f(x)
print(np.trapz(y, dx=dx))
```

Integrace pomocí Romberg

```python

print(integrate.romberg(f, a, b))
```
